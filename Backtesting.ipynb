```python
import pandas as pd
import numpy as np

# df_prices 是包含策略价格的DataFrame，索引为策略代码，列为日期
# df_selection 是包含策略选择标记的DataFrame，索引为策略代码，列为日期

# 计算涨跌幅的函数
def calculate_returns(df_prices, df_selection):
    # 获取所有日期
    dates = df_prices.columns
    # 初始化母策略的涨跌幅列表
    portfolio_returns = []

    # 遍历所有的日期，除了第一个因为没有前一个月的数据
    for i in range(1, len(dates)):
        current_date = dates[i]
        previous_date = dates[i-1]

        # 获取当前日期和前一个日期的价格
        current_prices = df_prices[current_date]
        previous_prices = df_prices[previous_date]

        # 计算涨跌幅
        returns = (current_prices - previous_prices) / previous_prices

        # 筛选出当前日期被选中的策略
        selected_strategies = df_selection[df_selection[current_date] == 1].index
        selected_returns = returns[selected_strategies]

        # 计算加权涨跌幅，假设每个策略的权重都是20%
        portfolio_return = (selected_returns * 0.20).sum()
        portfolio_returns.append(portfolio_return)

    # 计算总涨跌幅
    # total_return = np.prod(1 + portfolio_returns) - 1
    total_return = np.prod(1 + np.array(portfolio_returns)) - 1
    # 计算平均日涨跌幅
    average_daily_return = np.array(portfolio_returns).mean()

    # 计算夏普率，假设无风险利率为0
    risk_free_rate = 0
    excess_returns = np.array(portfolio_returns) - risk_free_rate
    sharpe_ratio = excess_returns.mean() / excess_returns.std() * np.sqrt(len(portfolio_returns))

    return total_return, average_daily_return, sharpe_ratio,pd.Series(portfolio_returns, index=pd.to_datetime(dates[1:]).strftime('%Y-%m'))
    # # 返回母策略的累计涨跌幅
    # return pd.Series(portfolio_returns, index=dates[1:])


df_prices = pd.read_csv('process_data.csv', index_col=0)
df_selection = pd.read_csv('top_5_policy.csv', index_col=0)

# 切片
test_df_prices = df_prices.loc[:, '2016-02':]
test_df_selection = df_selection.loc[:, '2016-02':]

train_df_prices = df_prices.loc[:, :'2016-01']
train_df_selection = df_selection.loc[:, :'2016-01']
# 调用函数计算母策略涨跌幅
# portfolio_returns = calculate_returns(df_prices, df_selection)

# print(portfolio_returns)

# 调用函数计算母策略涨跌幅及其他指标
total_return, average_daily_return, sharpe_ratio,day_return = calculate_returns(test_df_prices, test_df_selection)
train_total_return, train_average_daily_return, train_sharpe_ratio,train_day_return = calculate_returns(train_df_prices, train_df_selection)
# 打印结果
print("测试集的结果")
print(f"Total return: {total_return}")
print(f"Average daily return: {average_daily_return}")
print(f"Sharpe Ratio: {sharpe_ratio}")
print('-------------------')

print("训练集的结果")
print(f"Total return: {train_total_return}")
print(f"Average daily return: {train_average_daily_return}")
print(f"Sharpe Ratio: {train_sharpe_ratio}")
print('-------------------')
print(day_return)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates


plt.figure(figsize=(12, 6))
plt.plot(day_return.index, day_return.values, marker='o', linestyle='-', color='blue', markersize=8)

# 设置标题和标签
plt.title('Daily Portfolio Returns', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Returns', fontsize=14)

# 设置x轴日期格式
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=1))

# 美化图表
plt.grid(True, which='major', linestyle='--', linewidth=0.5, color='grey')
plt.axhline(0, color='red', linestyle='-', linewidth=1)  # 添加y=0的红色参考线

# 自动旋转日期标记以便阅读
plt.gcf().autofmt_xdate()

# 设置图例
plt.legend(['Portfolio Returns'], fontsize=12)

# 显示图表
plt.show()
```

```python
from pyecharts.charts import Line
from pyecharts import options as opts
from pyecharts.globals import ThemeType

# 将 pd.Series 转换为列表
dates = list(day_return.index) 
returns = list(day_return.values)  

# 创建一个折线图对象
line = Line(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))

# 添加X轴和Y轴数据
line.add_xaxis(dates)
line.add_yaxis("Portfolio Returns", returns, is_smooth=True)

# 设置全局配置项
line.set_global_opts(
    title_opts=opts.TitleOpts(title="Portfolio Returns"),
    xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False, axislabel_opts={"rotate": 45}),
    yaxis_opts=opts.AxisOpts(name="Returns"),
    tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross")
)

# 设置系列配置和样式
line.set_series_opts(
    label_opts=opts.LabelOpts(is_show=False),
    markline_opts=opts.MarkLineOpts(
        data=[opts.MarkLineItem(type_="average", name="平均值"),
              opts.MarkLineItem(y=0, name="基准线")],
    ),
    markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_="max", name="最大值"),
                                            opts.MarkPointItem(type_="min", name="最小值")])
)

# 渲染图表到文件，也可以直接渲染在Jupyter notebook中
line.render("day_return.html")
# 在Jupyter Notebook中渲染图表
line.render_notebook()
```

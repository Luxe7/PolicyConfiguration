```python
import pandas as pd
from datetime import datetime
```

提取每个月出现在表格中的第一天，并清洗数据

```python
df=pd.read_csv("processed_close_index.csv")
df = df.dropna()
df.set_index('Unnamed: 0', inplace=True)
df.index.name = None
df.columns = pd.to_datetime(df.columns)
dates = df.columns.tolist()
processed_months = set()

# 初始化列表来保存每个月的第一天
first_days = []

# 遍历排序后的日期列表
for date in dates:
    month_key = (date.year, date.month)  # 用年和月作为键
    if month_key not in processed_months:
        # 这是第一次遇到这个月，添加到结果列表中
        first_days.append(date)
        # 标记这个月已处理
        processed_months.add(month_key)
```

```python
process_df = df[first_days]
process_df
```

查看成长率

```python
result = (process_df - process_df.shift(axis=1))/process_df.shift(axis=1)
result = result.drop(result.columns[0],axis=1)
result
```

选取增长率前5的标签1，其余标签为0

```python
def process2Y(column):
    top_five = column.nlargest(5)
    return column.apply(lambda x: 1 if x in top_five.values else 0)

processed_result = result.apply(process2Y)
processed_result.columns = processed_result.columns.strftime('%Y-%m')
processed_result
```

```python
processed_result.to_csv('Count_Y.csv')
```

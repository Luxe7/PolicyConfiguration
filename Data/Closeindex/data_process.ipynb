```python
import pandas as pd
import os
```

```python
df=pd.read_csv("close_index.csv",dtype={'ticker': str})
print(df.shape)
df
```

```python
true_date=df[df['ticker']=='000015']['tradeDate'].tolist()
df_left = pd.DataFrame({'date':true_date})
len(true_date),df_left.head()
```

```python
# 按照 'Group' 列进行分组
grouped_df = df.groupby('ticker')
# 打印分组后的结果
for group_name, group_data in grouped_df:
    file_name=group_name+'_'+group_data['secShortName'].iloc[0]+".csv"
    result_df = pd.merge(df_left, group_data, left_on='date', right_on='tradeDate', how='left')[['date','closeIndex']] #为了确保日期是正确的
    result_df= result_df.set_index('date').T
    print(group_name,group_data.shape[0],file_name)
    # print(result_df)
    result_df.to_csv('./close_index_data/'+file_name,index=False)
```

## 将所有因子整合到一个文件中

```python
def get_file_df(file_path):
    df=pd.read_csv(file_path)
    return df,(df.apply(pd.Series.nunique) == 1).all()

def processed_data():
    df_dict={}
    data_dir_path=os.path.join(os.getcwd(), "close_index_data")
    data_file_name=os.listdir(data_dir_path)
    for file_name in data_file_name:
        file_path=os.path.join(data_dir_path, file_name)
        df,_=get_file_df(file_path)
        df_dict[file_name[:-4]]=pd.DataFrame(df.iloc[0]).T
    return df_dict

close_index_dict=processed_data()
```

```python
close_index_dict.keys()
```

```python
all_df = pd.DataFrame()
name_list=[]
for key,df in close_index_dict.items():
    name_list.append(key)
    all_df=pd.concat([all_df,df],axis=0,ignore_index=True)

all_df=all_df.set_index(pd.Index(name_list))
all_df
```

```python
all_df.to_csv('./processed_close_index.csv')
```

## 数据处理

```python
data_dir_path=os.path.join(os.getcwd(), "macro")
data_file_name=os.listdir(data_dir_path) #获取当前文件夹中所有文件名
data_file_name
```

```python
def get_all_df():
    # 获取列表，同时检查每一行的值是否相同
    df_dict={}
    data_dir_path=os.path.join(os.getcwd(), "macro")
    data_file_name=os.listdir(data_dir_path)
    for file_name in data_file_name:
        file_path=os.path.join(data_dir_path, file_name)
        df,Isunique=get_file_df(file_path)
        df_dict[file_name[:-4]]=df
        if not Isunique:
            print(f"{file_name}'s rows has different values")
    return df_dict



def get_file_df(file_path):
    
    df=pd.read_pickle(file_path)
    return df,(df.apply(pd.Series.nunique) == 1).all()
```

```python
tem=get_all_df()
tem.keys()
```

可以看到每一行的数据均为相同值，因此分析时可以将每一个DataFrame中保留一行，分析这一行的数据即可

```python
def processed_data():
    df_dict={}
    data_dir_path=os.path.join(os.getcwd(), "macro")
    data_file_name=os.listdir(data_dir_path)
    for file_name in data_file_name:
        file_path=os.path.join(data_dir_path, file_name)
        df,_=get_file_df(file_path)
        df_dict[file_name[:-4]]=pd.DataFrame(df.iloc[0]).T
    return df_dict
```

```python
data_processed=processed_data()
```

## 数据整合

```python
data_processed=processed_data()
```

```python
all_df = pd.DataFrame()
name_list=[]
for key,df in data_processed.items():
    name_list.append(key)
    all_df=pd.concat([all_df,df],axis=0,ignore_index=True)

all_df=all_df.set_index(pd.Index(name_list))
all_df
```

```python
all_df.T
```

```python
df=all_df.T
df=df.reset_index()
df.rename(columns={'index': 'Date'}, inplace=True)
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# 转置DataFrame
# df_transposed = df.pivot(index='Date', columns='Date').stack()

# 按月份合并并取平均值
df_monthly = df.groupby(df['Date'].dt.to_period("M")).mean()
df_monthly.T
```

```python
df_monthly.T.to_csv('./processed_macro_monthly.csv')
```
